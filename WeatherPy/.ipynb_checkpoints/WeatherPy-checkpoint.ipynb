The following is three observable trends from the following analysis:

1. By plotting the relationship between each city's latitude and the maximum temperature of the city, we can see a very close relationship. The city's that are closer to latitude 0, have significantly higher temperatures. This was an expected relationship. Latitude 0 is the equator, and therefore cities closer to the equator will have higher temperatures. What was not previously known, is that the northern hemisphere has a more linear relationship between latitude and maximum temperature than the sourthern hemisphere.

2. By plotting the relationship between each city's latitude and the humidity of the city, we can see no clear relationship between the variables. The linear regression plots show similar results. According to this analysis, there is no relationship between a city's latitude to it's humidity. This could be an area for further exploration. For example, is humidity more closer related to a city's proximity to body's of water? This will be explored graphically in the "VacationPy" notebook through a heat map of humidity.

3. Similarly to humidity, by plotting the relationship between each city's latitude and the cloudiness of the city, we can see no clear relationship between the variables. The linear regression plots show similar results. According to this analysis, there is no relationship between a city's latitude to it's cloudiness. This could be an area for further exploration. What makes some cities more cloudy than otheres?

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

```python
# Creating lists that will be appended in next section
city_names = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []
index_counter = 0
set_counter = 1
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
print(" Data Retrieval ")
print("-----------------------------")

base_url = "http://api.openweathermap.org/data/2.5/weather?"
units = "imperial"
query_url = f"{base_url}appid={weather_api_key}&units={units}&q="


# Appending data to lists for each city
for index, city in enumerate(cities, start = 1):
    try:
        response = requests.get(query_url + city).json()
        city_names.append(response["name"])
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        max_temp.append(response['main']['temp_max'])
        wind_speed.append(response["wind"]["speed"])
        if index_counter > 49:
            index_counter = 0
            set_counter = set_counter + 1
    
        else:
            index_counter = index_counter + 1
            
        print(f"Processing Record {index_counter} of Set {set_counter} : {city}") 
  
    except(KeyError, IndexError):
        print("City not found.")

print("Data Retrieval Complete")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# Creating data frame using data retrieved in lists
weather_df = pd.DataFrame({ 
                "City" : city_names,
                "Cloudiness" : cloudiness,
                "Country" : country,
                "Date" : date,
                "Humidity" : humidity,
                "Lat" : lat,
                "Lng" : lng,
                "Max Temp" : max_temp,
                "Wind Speed" : wind_speed
})
weather_df.head()
```

```python
#Desciring the weather_df
weather_df.describe()
```

## Inspect the data and remove the cities where the humidity > 100%.
----
Skip this step if there are no cities that have humidity > 100%. 

```python
# find cities that have humidity over 100%
humidity_100 = weather_df.loc[weather_df["Humidity"] > 100, :]

# Choose original data set if no cities have over 100% humidity
if humidity_100.empty == True:
    print("There are no cities with humidity values over 100%")
    print("Using original dataframe to continue with plotting...")
    
    # Grabbing on necessary variables from weather_df
    latitudes = weather_df["Lat"]
    temperatures = weather_df["Max Temp"]
    humidity = weather_df["Humidity"]
    cloudiness = weather_df["Cloudiness"]
    windspeed = weather_df["Wind Speed"]
    
     # Exporting the cleaned data to a CSV for later use
    clean_weather_data = weather_df
    clean_weather_data.to_csv("../output_data/clean_weather_data.csv", index=True, index_label="City ID", header=True)
else:
    # If cities are over 100% humidity, then get the index
    Humidity100Index = humidity_100.index
    
    # Drop cities that are over 100%
    clean_weather_data = weather_df.drop(Humidity100Index).reset_index(drop=True)
    clean_weather_data.head()
    
    # Get necessary variables from clean data
    latitudes = clean_weather_data["Lat"]
    temperatures = clean_weather_data["Max Temp"]
    humidity = clean_weather_data["Humidity"]
    cloudiness = clean_weather_data["Cloudiness"]
    windspeed = clean_weather_data["Wind Speed"]

    # Export the data to a CSV for later use
    clean_weather_data.to_csv("../output_data/clean_city_data.csv", index=True, index_label="City ID", header=True)
```

## Linear Regression

```python
# Creating a function for linear regression

def plot_linear_regression(x, y, ax=None, **kwargs):
    
    # using the axis that will be defined
    ax = ax
    
    # setting linear regression values and variables
    slope, intercept, rvalue, pvalue, stderror = linregress(x,y)  
    print(f"The R-Squared value is: {rvalue**2}")

    # setting equation for line
    regress_values = slope * x + intercept
    
    # creating the line equation
    line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
                                                          
    # plotting the line regression
    ax.plot(x, regress_values, color="red")
    
    # returning the line
    return ax, line_eq
```

```python
# Dividing the data by latitude 0
southern_hemisphere = clean_weather_data.loc[clean_weather_data["Lat"] < 0, :].copy().reset_index(drop=True)
northern_hemisphere = clean_weather_data.loc[clean_weather_data["Lat"] >= 0, :].copy().reset_index(drop=True)


# setting variables for northern vs southern hemispheres
north_latitudes = northern_hemisphere["Lat"]
north_temperatures = northern_hemisphere["Max Temp"]
north_windspeed = northern_hemisphere["Wind Speed"]
north_humidity = northern_hemisphere["Humidity"]
north_cloudiness = northern_hemisphere["Cloudiness"]


south_latitudes = southern_hemisphere["Lat"]
south_cloudiness = southern_hemisphere["Cloudiness"]
south_temperatures = southern_hemisphere["Max Temp"]
south_humidity = southern_hemisphere["Humidity"]
south_windspeed = southern_hemisphere["Wind Speed"]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Create the scatter plot
fig, ax = plt.subplots(figsize = (9,7))
plt.scatter(north_latitudes, north_temperatures, marker="o")

# Calling linear regression function
ax, line_eq = plot_linear_regression(north_latitudes, north_temperatures, ax)

# Adding label, titles and grid
plt.title("City Latitude vs. Max Temperature in the Northern Hemisphere", fontsize=14)
plt.ylabel("Max Temperature (F)", fontsize=10)
plt.xlabel("Latitude", fontsize=10)
plt.grid()

# Printing the regression line equation
plt.annotate(line_eq,(0,20),fontsize=14, color="red")

# Displaying and saving the plot
plt.savefig("../output_data/NorthernLatitudeVsTemp.png")
plt.show()
```

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
# Create the scatter plot
fig, ax = plt.subplots(figsize = (9,7))
plt.scatter(south_latitudes, south_temperatures, marker="o")

# Calling linear regression function
ax, line_eq = plot_linear_regression(south_latitudes, south_temperatures, ax)

# Adding label, titles and grid
plt.title("City Latitude vs. Max Temperature in the Southern Hemisphere", fontsize=12)
plt.ylabel("Max Temperature (F)", fontsize=10)
plt.xlabel("Latitude", fontsize=10)
plt.grid()

# Printing the regression line equation
plt.annotate(line_eq,(-35, 55),fontsize=14,color="red")

# Displaying and saving the plot
plt.savefig("../output_data/SouthernLatitudeVsTemp.png")
plt.show()
```

The two scatter plots with line regressions above explore the relationship between city latitude and maximum temperature for the northern and sourthern hemispheres. Both plots show the same relationship in terms of the trend of maximum temperature as latitude gets closer to zero. The northern hemisphere has a tighter trend and a higher r-squared value, which tells that the latitude and maximum temp has a more linear correlation in the northern hemisphere.
